## CREACIÓN DE UNA BASE DE DATOS

En este proyecto se va usar una base de datos para guardar la información, en lugar de por ejemplo un objeto guardado en formato de tipo ".pickle". Se va a usar el tipo de base de datos SQLite por su simplicidad y sencillez para usar en proyectos de este tipo.

Creamos la conexión:

```python
import sqlalchemy as sa

con = sa.create_engine('sqlite:///../../Datos_3/airbnb.db')
```

Creamos las tablas y cargamos los datos

```python
listings.to_sql('listings', con = con, if_exists = 'replace')
listings_det.to_sql('listings_det', con = con, if_exists = 'replace')
```
